```python
def haversine_np(lon1, lat1, lon2, lat2):
  lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])
  dlon = lon2 - lon1
  dlat = lat2 - lat1
  a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2
  c = 2 * np.arcsin(np.sqrt(a))
  km = 6367 * c
  return km
```

```python
df = pd.read_csv('nyc_taxis.csv')
```

```python
df.head()
```

# Feature Generation <a class="anchor" id="1"></a>

## Distance of trip <a class="anchor" id="2"></a>

```python
df['Distance_of_Trip'] = df.apply(lambda x: haversine_np(x.pickup_longitude, x.pickup_latitude, x.dropoff_longitude, x.dropoff_latitude), axis=1)
```

## Day of Week <a class="anchor" id="3"></a>

```python
df[["pickup_datetime", "dropoff_datetime"]] = df[["pickup_datetime", "dropoff_datetime"]].apply(pd.to_datetime)
```

```python
df['day_of_week_pick_up'] = df['pickup_datetime'].dt.day_name()
```

## Ave Speed of Trip <a class="anchor" id="4"></a>

```python
df['ave_speed_km_hr'] = df.apply(lambda x: x.Distance_of_Trip / (x.trip_duration*0.0002777), axis=1)
```

# Data Cleaning <a class="anchor" id="5"></a>

## Trip Duration <a class="anchor" id="6"></a>

```python
sns.histplot(df['trip_duration'],kde=False,bins=30)
```

```python
sns.boxplot(df['trip_duration'])
```

```python
df['trip_duration'].sort_values(ascending=False)
```

```python
df[df['trip_duration'] == 1].count()
```

```python
df[df['trip_duration'] > 86400]
```

From the above data exploration, it can be observed that several rows in the trip duration column exceed 24 hours and several trips have a duration of one second. These values are not intuitive in the context of this dataset and therefore the values were removed.  

```python
df.drop(df[(df['trip_duration'] > 86400) & (df['trip_duration'] == 1)].index, inplace = True)
```

## Passenger Count <a class="anchor" id="7"></a>

```python
df.passenger_count.value_counts()
```

```python
sns.countplot(x='passenger_count',data=df)
```

From the above distributions of the passenger count, it can be observed that 60 trips had no counted passengers. Furthermore, a few trips measured 7, 8 and 9 passengers. These rows will be removed due to their low occurrence and due to these values exceeding the average number passengers allowed in common taxis. 

```python
df=df[(df['passenger_count']!=0) & (df['passenger_count'] < 7)]
```

```python
df.passenger_count.value_counts()
```
